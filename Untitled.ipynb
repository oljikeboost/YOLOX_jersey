```python
import json
import mmcv
import cv2
import numpy as np
from yolox import inference
import matplotlib.pyplot as plt
```

```python
exp = inference.get_exp('exps/default/yolox_m_jersey.py', 'yolox-m')
exp.ckpt = './YOLOX_outputs/yolox_m_jersey/best_ckpt.pth.tar'
model = inference.create_model(exp)
model.cls_names = ['j']
```

```python
game = '2020_11_28_TexasSouthern_at_OklahomaState'
tracking = json.load(open('/home/ubuntu/oljike/data/demo_vids/'+game+'/'+game+'_tracking.json'))
video = mmcv.VideoReader('/home/ubuntu/oljike/data/demo_vids/'+game+'/'+game+'.mp4')
```

```python
w = video.width
h = video.height

all_crops = []
cnt = 0
for en, i in enumerate(tracking):
    if en%20!=0: continue
    
    frame = video.get_frame(int(i))
    
    for info in tracking[str(i)]:

        bbox = info.split(',')[1:5]

        bbox = [float(x) for x in bbox]
        bbox = [bbox[0] * w, bbox[1] * h, bbox[2] * w, bbox[3] * h]
        bbox = [int(x) for x in bbox]

        img_crop = frame[bbox[1]: bbox[1] + bbox[3], bbox[0]: bbox[0] + bbox[2], :]
        all_crops.append(img_crop)
    
    cnt += 1
    if cnt>100: break
```
